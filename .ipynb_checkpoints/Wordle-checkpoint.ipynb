```python
# Function that takes a guess and an answer as input, and returns the accuracy of the guess
# 2 is assigned to letters in the right place
# 1 is assigned to letters in the wrong place but not in the word
# 0 is for everything else
def getInformation(guess, answer):
    result = []
    for index in range(5):
        if(guess[index] in answer):
            if (guess[index] == answer[index]):
                result.append(2)
            else:
                result.append(1)
        else:
            result.append(0)
    return result

getInformation("crane", "cigar")
```

```python
# Function that reduces the remaining words based on a pattern and a guess
# Inputs: 
# remaining: All remaining possible answers
# guess: the word guessed
# pattern: the result pattern we got from the guessed word (array length 5 filled with 0s, 1s, and 2s)
def findRemainingWordsFromGuess(remaining, guess, pattern):
    remainingFiltered = []
    for word in remaining:
        # Can THIS word be the answer?
        if (pattern[0] == 0 and guess[0] in word and guess.count(guess[0]) == 1):
            continue
        if (pattern[1] == 0 and guess[1] in word and guess.count(guess[1]) == 1):
            continue
        if (pattern[2] == 0 and guess[2] in word and guess.count(guess[2]) == 1):
            continue
        if (pattern[3] == 0 and guess[3] in word and guess.count(guess[3]) == 1):
            continue
        if (pattern[4] == 0 and guess[4] in word and guess.count(guess[4]) == 1):
            continue
        if (pattern[0] == 1 and (guess[0] == word[0] or (guess[0] not in word))):
            continue
        if (pattern[1] == 1 and (guess[1] == word[1] or (guess[1] not in word))):
            continue
        if (pattern[2] == 1 and (guess[2] == word[2] or (guess[2] not in word))):
            continue
        if (pattern[3] == 1 and (guess[3] == word[3] or (guess[3] not in word))):
            continue
        if (pattern[4] == 1 and (guess[4] == word[4] or (guess[4] not in word))):
            continue
        if (pattern[0] == 2 and guess[0] != word[0]):
            continue
        if (pattern[1] == 2 and guess[1] != word[1]):
            continue
        if (pattern[2] == 2 and guess[2] != word[2]):
            continue
        if (pattern[3] == 2 and guess[3] != word[3]):
            continue
        if (pattern[4] == 2 and guess[4] != word[4]):
            continue
        else:
            remainingFiltered.append(word)
    return remainingFiltered
            
```

```python
# Function that formulates a guess based on the possible remaining guesses/answers, and which guess attempt we're on
def createGuess(possibleAnswers, possibleGuesses, guessNum):
    
    # Initialize the dictionary of all words. The value will represent how much we narrow down on average or total
    possibleGuesses = possibleGuesses if (guessNum <= 2 and len(possibleGuesses)<1000 and 
                                          len(possibleGuesses)>0 and len(possibleAnswers)>0) else possibleAnswers
    wordsRemaining = {key: [] for key in possibleGuesses}
    count = 0
    for answer in possibleAnswers:
        # we need to find what the pattern would be for each remaining possible word and
        # how much it narrows the list of possible answers down
        for guess in possibleGuesses:
            # get the patterns for each guess compared to this answer
            pattern = getInformation(guess, answer)
            
            # get the remaining lists for each guess/pattern
            remaining = findRemainingWordsFromGuess(possibleAnswers, guess, pattern)
            
            # add the length of remaining words (basically a measure of how much info we get) to the overall dictionary
            # of words
            wordsRemaining[guess].append(len(remaining)+1)
            
            #print(answer + " " + guess + ": " + str(pattern) + " => " + str(len(remaining)))
        count+=1
    for word in wordsRemaining:
        wordsRemaining[word] = sum(wordsRemaining[word])
    return min(wordsRemaining, key=wordsRemaining.get)
        
```

```python
def createGuessByScore(possibleAnswers, possibleGuesses, guessNum):
    
    if (guessNum > 2 or len(possibleGuesses) == 0):
        possibleGuesses = possibleAnswers
    
    # Initialize the dictionary of all words. The value will represent how much we narrow down on average or total
    scores = {key: [] for key in possibleGuesses}
    
    for answer in possibleAnswers:
        for guess in possibleGuesses:

            # get the patterns for each guess compared to this answer
            pattern = getInformation(guess, answer)
            
            # assign a score
            score = sum(pattern)
            
            scores[guess].append(score)

    for word in scores:
        scores[word] = sum(scores[word])
    return max(scores, key=scores.get)
```

```python
def getInformation(guess, answer):
    result = []
    for index in range(5):
        if(guess[index] in answer):
            if (guess[index] == answer[index]):
                result.append(2)
            else:
                result.append(1)
        else:
            result.append(0)
    return result

getInformation("raise", "cigar")
```

```python
# reduces the remaining words based on a patter guess
# returns remaining possible answers given a guess/pattern
def findRemainingWordsFromGuess(remaining, guess, pattern):
    remainingFiltered = []
    for word in remaining:
        # Can THIS word be the answer?
        if (pattern[0] == 0 and guess[0] in word and guess.count(guess[0]) == 1):
            # print(word + "1")
            continue
        if (pattern[1] == 0 and guess[1] in word and guess.count(guess[1]) == 1):
            #print(word + "2")
            continue
        if (pattern[2] == 0 and guess[2] in word and guess.count(guess[2]) == 1):
            #print(word + "3")
            continue
        if (pattern[3] == 0 and guess[3] in word and guess.count(guess[3]) == 1):
            #print(word + "4")
            continue
        if (pattern[4] == 0 and guess[4] in word and guess.count(guess[4]) == 1):
            #print(word + "5")
            continue
        if (pattern[0] == 1 and (guess[0] == word[0] or (guess[0] not in word))):
            #print(word + "6")
            continue
        if (pattern[1] == 1 and (guess[1] == word[1] or (guess[1] not in word))):
            #print(word + "7")
            continue
        if (pattern[2] == 1 and (guess[2] == word[2] or (guess[2] not in word))):
            #print(word + "8")
            continue
        if (pattern[3] == 1 and (guess[3] == word[3] or (guess[3] not in word))):
            #print(word + "9")
            continue
        if (pattern[4] == 1 and (guess[4] == word[4] or (guess[4] not in word))):
            #print(word + "10")
            continue
        if (pattern[0] == 2 and guess[0] != word[0]):
            #print(word + "11")
            continue
        if (pattern[1] == 2 and guess[1] != word[1]):
            #print(word + "12")
            continue
        if (pattern[2] == 2 and guess[2] != word[2]):
            #print(word + "13")
            continue
        if (pattern[3] == 2 and guess[3] != word[3]):
            #print(word + "14")
            continue
        if (pattern[4] == 2 and guess[4] != word[4]):
            #print(word + "15")
            continue
        else:
            remainingFiltered.append(word)
    return remainingFiltered
            
```

```python
def createGuess(possibleAnswers, possibleGuesses, guessNum):
    # Initialize the dictionary of all words. The value will represent how much we narrow down on average or total
    possibleGuesses = possibleGuesses if (guessNum <= 2 and len(possibleGuesses)<1000 and 
                                          len(possibleGuesses)>0 and len(possibleAnswers)>0) else possibleAnswers
    wordsRemaining = {key: [] for key in possibleGuesses}
    count = 0
    # supposed to be in allPossibleWords
    for answer in possibleAnswers:
        # we need to find what the pattern would be for each remaining possible word and
        # how much it narrows things down
        for guess in possibleGuesses:
            # get the patterns for each guess compared to this answer
            pattern = getInformation(guess, answer)
            # get the remaining lists for each guess/pattern
            remaining = findRemainingWordsFromGuess(possibleAnswers, guess, pattern)
            # add the length of remaining words (basically a measure of how much info we get) to the overall dictionary
            # of words
            wordsRemaining[guess].append(len(remaining)+1)
            #print(answer + " " + guess + ": " + str(pattern) + " => " + str(len(remaining)))
        count+=1
    for word in wordsRemaining:
        wordsRemaining[word] = sum(wordsRemaining[word])
    return min(wordsRemaining, key=wordsRemaining.get)
        
```

```python
def createGuessByScore(possibleAnswers, possibleGuesses, guessNum):
    
    if (guessNum > 2 or len(possibleGuesses) == 0):
        possibleGuesses = possibleAnswers
    
    # Initialize the dictionary of all words. The value will represent how much we narrow down on average or total
    scores = {key: [] for key in possibleGuesses}
    
    for answer in possibleAnswers:
        for guess in possibleGuesses:

            # get the patterns for each guess compared to this answer
            pattern = getInformation(guess, answer)
            
            # assign a score
            score = sum(pattern)
            
            scores[guess].append(score)

    for word in scores:
        scores[word] = sum(scores[word])
    return max(scores, key=scores.get)
```
